```python
def get_all_source(root):
    all_source_path = []
    for path, subdirs, files in os.walk(root):
        for name in files:
            #just source file
            if name[-5:] == '.java':
                all_source_path.append(str(pl.PurePath(path, name)))
    return sorted(all_source_path)

all_src_path = get_all_source(unzip_src_list[0])
len(all_src_path)
```

```python
import sys
import re
from nltk.stem.porter import *
ps = PorterStemmer()

def get_content(file_path):
    with open(file_path, 'rb') as f:
        text = f.read().decode(errors='replace')
        return text.replace('\ufffd',"")

def get_token(string, ignore_error= True):
    raw_tokens = javalang.tokenizer.tokenize(string, ignore_errors=ignore_error)
    tokens = []
    for t in raw_tokens:
        tokens.extend(re.sub( r"([A-Z\d]*[a-z\d]*)", r" \1", t.value).split()+[t.value])
    return [ps.stem(i.lower()) for i in tokens]
```

## Source code cleaning


```python

from sklearn.feature_extraction.text import CountVectorizer

stop_word = """abstract continue for new switch
assert default goto package synchronized
boolean do if private this
break double implements protected throw
byte else import public throws
case enum instanceof return transient
catch extends int short try
char final interface static void
class finally long strictfp volatile
const float native super while"""
black_list = [i+"[\s.:;\n\t]" for i in stop_word.split()]
black_list.append('\/\/.+\n') # comment // in java
black_list.append('0x[\w\d]+') # hex
black_list.append('<[\w/!\-.]+>') # html tag
black_list.append('["(\[{\s.:;,\n\t]-?\d+["dDfFLleE\s.,:;\n\t})\]]|[\s.,:;\n\t]\d+$|^\d+[\s.,:;\n\t]') # number
black_list.append(';[\d;]+') # number sequence

punc = """
( ) [ ]  ++ -- + - ! ~ * / % << >> >>>  < <= > >= instanceof == != & ^ | &&
|| ? : =  += -=  *= /= %=  <<= >>= >>>=  &= ^= |= ; { }  ,  . " \\ # ` @
_
"""
punc_list = punc.split()

def clean_source(string, black_list, punc_list = punc_list):
    for k in black_list:
        string = re.sub(k, ";", string)
    for p in punc_list:
        string = string.replace(p, " ")
    string = string.replace("'", "")
    return string

print(black_list)
print(punc_list)
```

```python
def get_content_bug_rp(bug_rp_df):
    """input: dataframe
    return: a series that contains summary concat with description of each bug report"""
    cleaned_summary = bug_rp_df['summary'].str.extract(r'Bug[ \d]+(.+)',expand = False)
    s = cleaned_summary + ". " + bug_rp_df['description'].fillna("")
    return s.tolist()

len(get_content_bug_rp(bug_rp_df))
```

## Tokenize

### Tokenize sourcecode

```python
import torchtext
from torchtext.data import get_tokenizer
import splitter

def all_token(src_files, black_list, punc_list = punc_list):
    "Construct vocab from all source files"
    tokens = []
    tokenizer = get_tokenizer("basic_english")
    for f in src_files:
        token = tokenizer(clean_source(get_content(f), black_list, punc_list))
        tokens.extend(token)
    #clean token, remove all cmt
    tokens = clean_source('  '.join(tokens), black_list).split()
    return tokens
```
